```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cities_path = './cities.csv'
cities_df = pd.read_csv(cities_path)
```

```python
cities_df.head()
```

```python
xaxis = cities_df['Lat']
yaxis = cities_df['Max Temp']
plt.title('Latitude vs Max Tempearture (°F)')
plt.xlabel('Latitude')
plt.ylabel('Temperature (°F)')
plt.xlim(-80,100)
plt.ylim(0,125)
plt.scatter(xaxis,yaxis,c='cornflowerblue',edgecolors='royalblue')
plt.savefig('./Visualisations/max_temp')
```

```python
xaxis = cities_df['Lat']
yaxis = cities_df['Cloudiness']
plt.title('Latitude vs Cloudiness')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness %')
plt.xlim(-80,100)
plt.ylim(0,100)
plt.scatter(xaxis,yaxis,c='cornflowerblue',edgecolors='royalblue')
plt.savefig('./Visualisations/cloudiness')
```
